Reto 4
Encapsula los comportamientos anteriores usando funciones

```python
pasos_x = int(input("Ingrese los pasos que avanzará la tortuga ")) #se definen las variable con tipo entero para que el usuario ingrese los valores
pasos_y = int(input("Ingrese los pasos que bajará la tortuga? "))

def adelante(pasos): #esta función imprime los pasos horizontales
    pasos_x = pasos
    print("-" * pasos + ">")


def abajo(pasos): #esta función imprime los pasos verticales iniciando desde la ultima posición horizontal
    for i in range(pasos):
        print(" " * pasos_x + "|")

    print(" " * pasos_x + "V")

adelante(pasos_x)
abajo(pasos_y)
```

Reto 5


```python
import turtle #Se importa la librería turtle

tortuga = turtle.Turtle() #Se crea un objeto tortuga
tortuga.speed(1)

# Funciones de movimiento

def adelante(pasos):
    tortuga.forward(pasos)

def abajo(pasos):
    tortuga.right(90)
    tortuga.forward(pasos)
    tortuga.left(90)

# Entrada de datos del usuario
peldaños = int(input("Ingrese la cantidad de peldaños: "))
pasos_x = int(input("Ingrese los pasos horizontales por peldaño: "))
pasos_y = int(input("Ingrese los pasos verticales por peldaño: "))


# Dibujo de la escalera

for i in range(peldaños):
    adelante(pasos_x)  # tramo horizontal
    abajo(pasos_y)       # tramo vertical


tortuga.done()
```
